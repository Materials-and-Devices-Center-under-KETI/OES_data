```python
import pandas as pd
c10_p1 = pd.read_csv('file/process2/N2_10th_pattern1.csv', index_col=0)
c10_p2 = pd.read_csv('file/process2/N2_10th_pattern2.csv', index_col=0)
c10_p3 = pd.read_csv('file/process2/N2_10th_pattern3.csv', index_col=0)
dfs = [c10_p1, c10_p2, c10_p3]
c10 = pd.concat(dfs, axis=1, ignore_index=True)
c10.columns = ['pattern1','pattern2','pattern3']
c10
```

```python
ref = pd.read_excel('file/etc/emission_spectrum.xlsx')
print(len(ref['name'].unique()))
ref.head()
```

```python
import numpy as np

# module to return names of candidate molecules
def getNames(wavelength, margin=0.5):
    if wavelength == np.nan:
        return np.nan
    candidates = [] # set
    for idx, row in ref.iterrows():
        if abs(row['wavelength'] - wavelength) <= margin:
            candidates.append(row['name'])
    return ''.join(map(lambda x: str(x)+',',set(candidates)))

getNames(294.64)
```

```python
c10['pattern1_name'] = c10['pattern1'].apply(getNames)
c10['pattern2_name'] = c10['pattern2'].apply(getNames)
c10['pattern3_name'] = c10['pattern3'].apply(getNames)
```

```python
c10.to_csv('file/process2/N2_pattern_name.csv', index=False)
```

```python
print(c10['pattern1_name'].unique())
print(c10['pattern2_name'].unique())
print(c10['pattern3_name'].unique())
```

```python
def extractNames(arr):
    names = []
    for i in arr:
        names.extend(i.split(','))
    while '' in names:
        names.remove('')
    return set(names)

print(len(extractNames(c10['pattern1_name'].unique())), extractNames(c10['pattern1_name'].unique()))
print(len(extractNames(c10['pattern2_name'].unique())), extractNames(c10['pattern2_name'].unique()))
print(len(extractNames(c10['pattern3_name'].unique())), extractNames(c10['pattern3_name'].unique()))
```
